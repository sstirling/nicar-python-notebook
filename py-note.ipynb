```python
merged.groupby(["contributor_lastname", "contributor_firstname","committee_position"]).amount.sum().reset_index().sort_values("amount", ascending=False)
```

```python
merged.groupby(["contributor_employer","committee_position"]).amount.sum().reset_index().sort_values("contributor_employer")
```

```python
top_supporters = support.groupby(["contributor_firstname", "contributor_lastname"]).amount.sum().reset_index().sort_values("amount", ascending=False).head(10)
```

```python
chart = top_supporters.amount.plot.barh()
chart.set_yticklabels(top_supporters.contributor_lastname)
```

```python
top_supporters["contributor_fullname"] = top_supporters.contributor_firstname + " " + top_supporters.contributor_lastname
```

```python
chart = top_supporters.amount.plot.barh()
chart.set_yticklabels(top_supporters.contributor_fullname)
```

```python
top_supporters.to_csv("top-supporters.csv", index=False)
```

```python
top_opponents = oppose.groupby(["contributor_firstname", "contributor_lastname"]).amount.sum().reset_index().sort_values("amount", ascending=False).head(10)
```

```python
top_opponents["contributor_fullname"] = top_opponents.contributor_firstname + " " + top_opponents.contributor_lastname
```

```python
chart = top_opponents.amount.plot.barh()
chart.set_yticklabels(top_opponents.contributor_fullname)
```

```python
top_opponents.to_csv("top-opponents.csv", index=False)
```
